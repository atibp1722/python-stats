```python
import pandas as pd
```

```python
data=pd.read_csv('Car_sales.csv')
data.head(10)
```

```python
data.describe()
```

```python
import numpy as np

np.ptp(data['Price_in_thousands'])
np.percentile(data['Sales_in_thousands'],75)
```

```python
from scipy.stats import kurtosis
from scipy.stats import skew
from scipy.stats import zscore

print("Karl-Pearson kurtosis,",kurtosis(data['Sales_in_thousands'],fisher=False))
print("Fisher kurtosis",kurtosis(data['Sales_in_thousands'],fisher=True))
```

```python
skew(data['Sales_in_thousands'])
```

```python
zscore(data['Sales_in_thousands'])
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(6,6))
plt.boxplot(data['Sales_in_thousands'])
plt.show()
```
