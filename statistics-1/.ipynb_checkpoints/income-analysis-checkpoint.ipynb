```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('Customers.csv')
data.head(10)
```

```python
profession_group=data.groupby('Profession')
profession_group.get_group('Lawyer').head(10)
```

```python
data.isnull()
```

```python
data['Profession'].unique()
```

```python
print(data['Age'].mean())
print(data['Annual Income ($)'].mean())
print(data['Spending Score (1-100)'].mean())
print(data['Work Experience'].mean())
print(data['Family Size'].mean())
print(data['Annual Income ($)'].mode())
```

```python
print(data['Spending Score (1-100)'].var())
print(data['Spending Score (1-100)'].std())
```

```python
from scipy.stats import skew
from scipy.stats import zscore

print(skew(data['Annual Income ($)']))
print(skew(data['Spending Score (1-100)']))
```

```python
score=np.abs(zscore(data[['Annual Income ($)','Spending Score (1-100)']]))
score
```

```python
np.where(score>2)
```
