```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.kdeplot(data['price'])
```

```python
import scipy.stats as stats
import pylab

stats.probplot(data['price'],plot=pylab)
```

```python
stats.probplot(data['citympg'],plot=pylab)
```

```python
def plotNorm(data,feature):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    sns.kdeplot(data[feature])
    plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()
```

```python
plotNorm(data,'price')
```

```python
data['price_log']=np.log(data['price'])
plotNorm(data,'price_log')
```

```python
data['price_recip']=1/(data['price'])
plotNorm(data,'price_recip')
```

```python
data['price_sqrt']=np.sqrt(data['price'])
plotNorm(data,'price_sqrt')
```

```python
data['price_exp']=data['price']**(1/1.2)
plotNorm(data,'price_exp')
```

```python
data['price_boxcox'],parameters=stats.boxcox(data['price'])
plotNorm(data,'price_boxcox')
```
