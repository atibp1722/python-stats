```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
languages=['Python','Ruby','Rust','SQL','Java','C++','Swift']
pay=[87000,91000,110000,70000,77000,81000,86000]
popularity=[28,12,7,19,23,8,3]
```

```python
plt.pie(popularity,labels=languages,autopct='%1.1f%%')
plt.title('Languages by Popularity')
plt.show()
```

```python
plt.hist(pay,edgecolor='r')
plt.xlabel("Average Pay")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
year=[2015,2016,2017,2018,2019]
epl=[75,89,69,80,100]
laliga=[77,65,142,120,60]
seriea=[90,50,55,60,75]
```

```python
plt.plot(year,epl,label='EPL',color='red')
plt.plot(year,laliga,label='La LIga',color='green')
plt.plot(year,seriea,label='Serie A',color='blue')
plt.xlabel("Year")
plt.ylabel("Fees (in millions)")
plt.legend()
plt.show()
```
