```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from scipy.stats import bernoulli

data=bernoulli.rvs(size=1000,p=0.4)
sns.distplot(data)
plt.xlabel("Bernoulli")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.stats import uniform

n=1000
x=0
y=10
data=uniform.rvs(size=n,loc=x,scale=y)
sns.distplot(data,bins=100)
plt.xlabel("Uniform")
plt.ylabel("Frquency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

num=[1,4,56,76,12,10]
means=[]

for i in num:
    np.random.seed(0)
    x=[np.mean(np.random.randint(-50,50,i)) for _ in range(1000)]
    means.append(x)
    
k=0

fig,ax=plt.subplots(2,2,figsize=(12,12))
for i in range(0,2):
    for j in range(0,2):
        ax[i,j].hist(means[k],10,density=True)
        ax[i,j].set_title(label=num[k])
        k+=1
plt.show()
```
