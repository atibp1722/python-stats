```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('bank.csv')
data.head(5)
```

```python
data.describe()
```

```python
data.groupby('job')['housing'].value_counts()
```

```python
pd.crosstab(data['job'],data['housing'])
```

```python
pd.crosstab(data['education'],data['loan'],normalize='index')
```

```python
sns.scatterplot(x='balance',y='age',data=data,hue='loan')
```

```python
sns.lmplot(x='balance',y='age',data=data,hue='loan')
```

```python
sns.lmplot(x='balance',y='age',data=data,col='loan')
```

```python
sns.pairplot(data)
```

```python
c=data.corr()
plt.figure(figsize=(8,8))
sns.heatmap(c,annot=True)
```
