# HOUSE PRICE PREDICTION

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

sizes = [1500, 2000, 1200, 1800, 2400]
prices = [200000, 250000, 180000, 220000, 300000]

plt.scatter(sizes, prices)
plt.xlabel('Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Prepare the data
sizes = [[1500], [2000], [1200], [1800], [2400]]
prices = [200000, 250000, 180000, 220000, 300000]

# Create and fit the model
model = LinearRegression()
model.fit(sizes, prices)

# Predict the price for a new house with a size of 1800 sq. ft.
new_house_size = [[1800]]
predicted_price = model.predict(new_house_size)

print("Predicted price:", predicted_price)
```

```python
# Generate points along the x-axis for plotting the line
x = np.linspace(1000, 3000, 100)
x = x.reshape(-1, 1)

# Predict prices for the generated points
y = model.predict(x)

# Plot the scatter plot and the best-fit line
plt.scatter(sizes, prices)
plt.plot(x, y, color='red', label='Best-fit line')
plt.xlabel('Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```
