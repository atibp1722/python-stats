```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('bank.csv')
data.head()
```

```python
x=data['age']
y=data['balance']

plt.scatter(x,y,color='red')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

x=x.values.reshape(-1,1)
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.predict([[20]])
```

```python
predict_val=lr.predict(x)
plt.scatter(x,y,color='black',label='Data Points')
plt.plot(x,predict_val,color='green',label='Line of Regression')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(y,predict_val)
```

# Assumptions

```python
sns.pairplot(data,x_vars=['age','balance','duration'],y_vars='campaign')
```

```python
predicted_val=lr.predict(x)
residuals=(y-predicted_val)

mean_residual=np.mean(residuals)
print('Mean residual',mean_residual)
```

```python
plot=sns.scatterplot(predicted_val,residuals)
plot=sns.lineplot([0,1000],[0,0])
plt.xlabel('Predicted Values')
plt.ylabel('Residual Values')
plt.show()
```
