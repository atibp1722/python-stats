```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('insurance.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
cats=data.select_dtypes(include=['object'])
cats
```

```python
cats_conv=ohe.fit_transform(cats).toarray()
cats_conv
```

```python
cats_encoded=pd.DataFrame(cats_conv)
cats_encoded
```

```python
data.drop(columns=data.columns,inplace=True)
```

```python
new_data=pd.concat([cats_encoded,data],ignore_index=True,axis=1)
```

```python
X=new_data.iloc[:,:-1]
y=new_data.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred=lr.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
