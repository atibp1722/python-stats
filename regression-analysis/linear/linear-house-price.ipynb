```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('houseprices.csv')
data.head()
```

```python
import statsmodels.api as sms

X=data['Bedrooms']
Y=data['Price']
```

```python
X=sms.add_constant(X)
lr=sms.OLS(Y,X).fit()

lr.predict(X)
```

```python
lr.summary()
```

```python
y_pred=4.248e+04*X['Bedrooms']

sns.scatterplot(x=X['Bedrooms'],y=Y)
sns.lineplot(x=X['Bedrooms'],y=y_pred,color='green')
plt.show()
```
