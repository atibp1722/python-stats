```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data=pd.read_csv('real_estate_price_size_year_view.csv')
raw_data
```

```python
data=raw_data.copy()
data['view']=data['view'].map({'Sea view':1,'No sea view':0})
data
```

```python
y=data['price']
x1=data[['size', 'view']]
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data['size'],data['price'],c=data['view'], cmap='RdYlGn_r')
yhat_no=77480+218.75*data['size']
yhat_yes=133240+218.75*data['size']
fig=plt.plot(data['size'], yhat_no, lw=2, c='green')
fig=plt.plot(data['size'], yhat_yes, lw=2, c='red')
plt.xlabel('Size', fontsize=20)
plt.xlabel('Price', fontsize=20)
plt.show()
```
