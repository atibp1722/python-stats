```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def linear(x,w,b):
    return w*x+b

def logistic(x,w,b):
    return 1/(1+np.exp(-linear(x,w,b)))
```

```python
linear(6,4,1)
```

```python
logistic(6,4,1)
```

```python
w=-2.8
b=-4
x=np.linspace(-5,5,50)
y=logistic(x,w,b)
plt.plot(x,y)
plt.title('Logistic Plot at w={} and b={}'.format(w,b))
plt.show()
```

# Logistic Model 

```python
from sklearn.datasets import make_classification

X,y=make_classification(n_features=2,n_redundant=0,n_informative=2,random_state=1,n_clusters_per_class=1)

import seaborn as sns
sns.scatterplot(X[:,0],X[:,1],hue=y)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def  lossFunction(y,y_hat):
    loss=-np.mean(y*(np.log(y_hat))-(1-y)*np.log(1-y_hat))
    return loss

def gradient(X,y,y_hat):
    m=X.shape[0]
    
    dw=(1/m)*np.dot(X.T,(y_hat-y))
    db=(1/m)*np.sum((y_hat-y))
    
    return dw,db

def decision(X,w,b):
    x1=[min(X[:,0]),max(X[:,0])]
    m=-w[0]/w[1]
    c=-b/w[1]
    x2=m*x1+c
    
    fig=plt.figure(figsize=(6,6))
    plt.plot(X[:,0][y==0],X[:,1][y==0],'g^')
    plt.plot(X[:,0][y==1],X[:,1][y==1],'bs')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Limit')
    plt.plot(x1,x2,'y-')
    
def normalize(X):
    m,n=X.shape
    for i in range(n):
        X=(X-X.mean(axis=0))/X.std(axis=0)
    return X
```

```python
def train(X,y,bs,epoch,lr):
    m,n=X.shape
    w=np.zeros((n,1))
    b=0
    y=y.reshape(m,1)
    x=normalize(X)
    losses=[]
    
    for i in range(epoch):
        for i in range((m-1)//bs+1):
            start=i*bs
            end=start+bs
            xb=X[start:end]
            yb=y[start:end]
            \
            y_hat=sigmoid(np.dot(xb,w)+b)
            dw,db=gradient(xb,yb,y_hat)
            
            w-=lr*dw
            b-=lr*db
            
            l=lossFunction(y,sigmoid(np.dot(X,w)+b))
            losses.append(l)
            
            return w,b,losses
```

```python
def predict(X):
    x=normalize(X)
    pred=sigmoid(np.dot(X,w)+b)
    pred_c=[]
    pred_c=[1 if i>0.5 else 0 for i in pred]
    return np.array(pred_c)

w,b,l=train(X,y,bs=100,epoch=500,lr=0.01)
decision(X,w,b)
```

```python
y_predict=predict(X)

from sklearn.metrics import accuracy_score
accuracy_score(y,y_predict)
```
