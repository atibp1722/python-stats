```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
samples=200
age=np.random.randint(21,60,size=samples)
gender=np.random.choice([1,0],size=samples)
experience=np.random.randint(0,25,size=samples)
current_salary=np.random.randint(20000,200000,size=samples)
inflation=np.random.uniform(0.01,0.25,size=samples)
education=np.random.randint(0,4,size=samples)

actual_salary=current_salary/(1+inflation)
#print(actual_salary)
```

```python
median_salary=np.median(actual_salary)
target=(actual_salary>median_salary).astype(int)
# print(median_salary)
# print(target)
```

```python
from scipy.special import expit

def logisticRegression(f,w):
    z=np.dot(f,w)
    return expit(z)

features=6
w=np.random.randn(features)

learn_rate=0.01
iterations=1000
```

```python
for _ in range(iterations):
    predict=logisticRegression(np.column_stack((age,gender,experience,current_salary,inflation,education)),w)
    errors=target-predict
    gradient=np.dot(np.column_stack((age,gender,experience,current_salary,inflation,education)).T,errors)
    w+=learn_rate*gradient
```

```python
fig=plt.figure(figsize=(8,4), dpi= 100, facecolor='w', edgecolor='k')

plt.scatter(age,experience,c=target,cmap='coolwarm',marker='X',label='Salary below median value')
plt.xlabel('Age')
plt.ylabel('Experience')
plt.legend()
plt.title('Salary Plot')
plt.colorbar(label='Salary Classification')
plt.show()
```
